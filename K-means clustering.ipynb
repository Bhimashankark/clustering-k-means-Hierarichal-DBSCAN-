## K Means Clustering

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,n_features=2,random_state=23)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
## now performing train_test_split for training the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## lets plot the elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
y_labels=kmeans.fit_predict(X_train)
```

```python
kmeans.labels_
```

```python
y_labels
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
#silhouette_score
from sklearn.metrics import silhouette_score
```

```python
silhoute_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X_train)
    score=silhouette_score(X_train,labels=kmeans.labels_)
    silhoute_coefficients.append(score)
```

```python
silhoute_coefficients
```

```python
## plotting silhouette score
plt.plot(range(2,11),silhoute_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
